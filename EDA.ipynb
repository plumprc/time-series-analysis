```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('GunPoint/GunPoint_TRAIN.tsv', sep='\t', header=-1)
df.head()
```

```python
traces = df.iloc[10:18].values
```

```python
plt.figure(figsize=(10, 4))
for index, trace in enumerate(traces):
    plt.subplot(2, 4, index + 1)
    plt.plot(trace[1:])
    plt.axis('off')
    plt.title('class=' + str(trace[0]))

plt.show()
```
