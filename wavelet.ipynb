```python
import numpy as np
from matplotlib import pyplot as plt
import numpy.ma as ma
```

```python
def fft(y):
    comp = np.fft.fft(y)
    comp = np.sqrt(comp.real ** 2 + comp.imag ** 2)
    return comp
```

```python
x = np.linspace(0, 5, 100)
y = np.exp(-x) * np.cos(5 * x) + np.random.normal(0, 0.005, 100)
```

```python
plt.plot(x, y)
plt.show()
```

```python
f = fft(y)
plt.plot(x, f)
plt.show()
```

```python
y_crop = y[20:80]
f = fft(y_crop)
plt.plot(x[20:80], f)
plt.show()
```

```python
mask = np.random.binomial(1, 0.5, 100)
y_mask = ma.masked_array(y, mask)
f = fft(y_mask)
plt.plot(f)
plt.show()
```
